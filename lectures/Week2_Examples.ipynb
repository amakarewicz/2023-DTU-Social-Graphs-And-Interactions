### Network notation

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
G1 = nx.Graph() # undirected network
G1.add_nodes_from(["a", "b", "c", "d", "g"])
G1.add_edges_from([("a", "b"), ("b", "c"), ("a", "c"), ("c", "d"), ("c", "g")])
```

```python
plt.figure()
nx.draw(G1, pos=nx.spring_layout(G1), with_labels=True)
```

```python
print("A = ", nx.to_numpy_array(G1)) # adjacency matrix
```

```python
G2 = nx.DiGraph() # directed network
G2.add_nodes_from(["a", "b", "c", "d", "g"])
G2.add_edges_from([("a", "b"), ("b", "a"), ("b", "c"), ("a", "c"), ("c", "d"), ("c", "g")])

plt.figure()
nx.draw(G2, pos=nx.spring_layout(G2), with_labels=True)
plt.show()
```

```python
print("A = ", nx.to_numpy_array(G2)) # directed so not necessarily symmetric
```

```python
G = nx.Graph()
G.add_edge("a", "b", weight=0.6)
G.add_edge("b", "c", weight=0.6)
G.add_edge("a", "c", weight=0.2)
G.add_edge("c", "d", weight=0.1)
G.add_edge("c", "e", weight=0.7)

pos = nx.spring_layout(G)

estrong = [(u,v) for (u,v,d) in G.edges(data=True) if d["weight"] > 0.5]
eweak = [(u,v) for (u,v,d) in G.edges(data=True) if d["weight"] <= 0.5]

nx.draw_networkx_nodes(G, pos, node_size=400)

nx.draw_networkx_edges(G, pos, edgelist=estrong, width=6)
nx.draw_networkx_edges(G, pos, edgelist=eweak, width=2, alpha=0.5, edge_color="b", style="dashed")

nx.draw_networkx_labels(G, pos, font_size=15, font_family="sans-serif")
edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels)

ax = plt.gca()
ax.margins(0.08)

plt.axis("off")
plt.tight_layout()

# nx.draw(G)
```

```python
print("A = ", nx.to_numpy_array(G)) # weights instead of ones
```

### Random networks

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# n - number of nodes
# p - probability that any one edge exists

# create nodes first, then take a pair and toss a coin
# with prob p we place an edge between them
```

```python
G = nx.fast_gnp_random_graph(n=50, p=0.15)
nx.draw(G,
        pos=nx.spring_layout(G),
        node_size=40,
        node_color='C0',
        edge_color='C1'
        )
```
