```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
GER = nx.fast_gnp_random_graph(n=10000, p=6/200)
# nx.draw(GER, node_size=20, node_color='C0', edge_color='C1')

ER_degree_freq = nx.degree_histogram(GER)
ER_degrees = range(len(ER_degree_freq))

plt.figure()
plt.plot(ER_degrees, ER_degree_freq, 'k')
plt.xlim(0, 700)
plt.xlabel("Degree")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
fig, ax = plt.subplots(1,2,figsize=(3.6*3, 3.6*1.5))

GER = nx.fast_gnp_random_graph(n=200, p=6/200)
nx.draw(GER, node_size=20, node_color='C0', edge_color='C1', ax=ax[0])

GBA = nx.barabasi_albert_graph(n=200, m=3)
nx.draw(GBA, node_size=20, node_color='C0', edge_color='C1', ax=ax[1])
plt.show()
```

```python
ER_degree_freq = nx.degree_histogram(GER)
ER_degrees = range(len(ER_degree_freq))

BA_degree_freq = nx.degree_histogram(GBA)
BA_degrees = range(len(BA_degree_freq))
```

```python
fig, ax = plt.subplots(3,2,figsize=(3.6*3,3.6*3))

ax[0,0].set_title("Erdos-Renyi")
nx.draw(GER, node_size=20, node_color='C0', edge_color='C1', ax=ax[0,0])

ax[0,1].set_title("Barabasi-Albert")
nx.draw(GBA, node_size=20, node_color='C0', edge_color='C1', ax=ax[0,1])

ax[1,0].plot(ER_degrees, ER_degree_freq, 'ko-')
ax[1,0].set_xlabel("Degree")
ax[1,0].set_ylabel("Count")

ax[1,1].plot(BA_degrees, BA_degree_freq, 'ko-')
ax[1,1].set_xlabel("Degree")
ax[1,1].set_ylabel("Count")

ax[2,0].plot(ER_degrees, ER_degree_freq, 'ko-')
ax[2,0].set_xlabel("Degree")
ax[2,0].set_ylabel("Count")
ax[2,0].set_xscale("log")
ax[2,0].set_yscale("log")

ax[2,1].plot(BA_degrees, BA_degree_freq, 'ko-')
ax[2,1].set_xlabel("Degree")
ax[2,1].set_ylabel("Count")
ax[2,1].set_xscale("log")
ax[2,1].set_yscale("log")
```

```python
GER = nx.fast_gnp_random_graph(n=10000, p=6/10000)
# nx.draw(GER, node_size=20, node_color='C0', edge_color='C1', ax=ax[0])

GBA = nx.barabasi_albert_graph(n=10000, m=3)
# nx.draw(GBA, node_size=20, node_color='C0', edge_color='C1', ax=ax[1])

ER_degree_freq = nx.degree_histogram(GER)
ER_degrees = range(len(ER_degree_freq))

BA_degree_freq = nx.degree_histogram(GBA)
BA_degrees = range(len(BA_degree_freq))
```

```python
fig, ax = plt.subplots(2,2,figsize=(3.6*2,3.6*3))

ax[0,0].plot(ER_degrees, ER_degree_freq, 'ko-')
ax[0,0].set_xlabel("Degree")
ax[0,0].set_ylabel("Count")

ax[0,1].plot(BA_degrees, BA_degree_freq, 'ko-')
ax[0,1].set_xlabel("Degree")
ax[0,1].set_ylabel("Count")

ax[1,0].plot(ER_degrees, ER_degree_freq, 'ko-')
ax[1,0].set_xlabel("Degree")
ax[1,0].set_ylabel("Count")
ax[1,0].set_xscale("log")
ax[1,0].set_yscale("log")

ax[1,1].plot(BA_degrees, BA_degree_freq, 'ko-')
ax[1,1].set_xlabel("Degree")
ax[1,1].set_ylabel("Count")
ax[1,1].set_xscale("log")
ax[1,1].set_yscale("log")
```

### Barabasi-Albert model

```python
# networks grow (new nodes come and attach to existing nodes)
# new nodes prefer to attach to already-popular nodes (preferential attachment)
# new node attached to node i with probability proportional to its degree
```

```python
# step 1
G = nx.Graph()
G.add_edges_from([(0,1)])
G.add_node(2)
nx.draw(G)
```
